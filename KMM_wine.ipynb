```python
import numpy as np
filename = 'wine.data'
f = open(filename,'r')
dataset = []
for x in f:
    temp_ls = []
    temp_s = ''
    for ch in x:
        if (ch >='0' and ch <= '9') or ch == '.':
            temp_s += ch
        elif ch == ',' or ord(ch) == 10:
            temp_ls.append(float(temp_s))
            temp_s = ''
        else:
            break
    
    dataset.append(temp_ls)
# print(dataset)
```

```python
import random

A = []
m = 10
n = len(dataset)
K = 5
h = 10

   #Under consideration

randomlist = random.sample(range(0, len(dataset)-1), m)
for index in randomlist:
    A.append(dataset[index])
    
# print(A)
```

```python
#This cell is for initializing S. Do not delete this cell
d_x = np.zeros((n,m))
for i in range(n):
    for j in range(m):
        d_x[i][j] = np.linalg.norm(dataset[i] - A[j]) ** 2
    
    d_x[i].sort()
```

```python
def update_d_x(A):
    new_d_x = np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            new_d_x[i][j] = np.linalg.norm(dataset[i] - A[j]) ** 2

        new_d_x[i].sort()
    
    return new_d_x
```

```python
def update_S(dist_vector):
    
#     for i in range(len(dist_vector)):
#         temp = dist_vector[i]
#         dist_vector[i] = temp.sort()
    
    S = np.zeros((n,m))
    for i in range(n):
        for j in range(m):
            if j < K:
                S[i][j] = (dist_vector[i][K+1] - dist_vector[i][j])/((K*dist_vector[i][K+1]) - sum(dist_vector[i][:K]))
            else:
                S[i][j] = 0
            
            if S[i][j] < 0:
                print('i:' + str(i) + ' j:' + str(j))
    return S
```

```python
def update_P(S):
    n_zeros_array = np.zeros((n,n))
    m_zeros_array = np.zeros((m,m))
    S_transpose = S.transpose()

    P = np.block([[n_zeros_array,S],[S_transpose,m_zeros_array]])
    return P
```

```python
def update_degrees(P):
    D_U = np.zeros((n,n))
    D_V = np.zeros((m,m))

    for i in range(P.shape[0]):
        if i < n:
            D_U[i][i] = sum(P[i])
        else:
            D_V[i-n][i-n] = sum(P[i])

    D = np.block([[D_U,np.zeros((n,m))],[np.zeros((m,n)),D_V]])
    
    return (D,D_U,D_V)
```

```python
from scipy.linalg import sqrtm
import math

def update_S_bar(D_U,D_V):
# temp_u = np.linalg.inv(sqrtm(D_U))
# np.linalg.det(D_V)
# temp_v = np.linalg.inv(sqrtm(D_V))
# temp_v

    temp_u = np.zeros((n,n))
    temp_v = np.zeros((m,m))

    for i in range(n):
        if D_U[i][i] != 0:
            temp_u[i][i] = 1/math.sqrt(D_U[i][i])
        else:
            temp_u[i][i] = 0

    for i in range(m):
        if D_V[i][i] != 0:
            print(D_V[i][i])
            temp_v[i][i] = 1/math.sqrt(D_V[i][i])
        else:
            temp_v[i][i] = 0

    S_bar = temp_u @ S @ temp_v
    
    return S_bar
```

```python
def update_F(S_bar):
    U, dig, V_transpose = np.linalg.svd(S_bar)
    V = V_transpose.transpose()
    
    final_U = (1/math.sqrt(2))*U[:K].reshape(n,K)
    final_V = (1/math.sqrt(2))*V[:K].reshape(m,K)
    F = np.block([[final_U],[final_V]])
    
    return F
```

```python
def update_d_i_bar(F,D,i):
    d_i_bar = np.array([])
    for j in range(m):
        temp = np.linalg.norm(dataset[i] - A[j]) ** 2
#         print(d_i_bar)
        if D[i][i] != 0 and D[n+j][n+j] != 0:
            temp += h*np.linalg.norm((F[i]/math.sqrt(D[i][i])) - (F[n+j]/math.sqrt(D[n+j][n+j])))
        
        d_i_bar = np.append(d_i_bar,[temp])
    
    d_i_bar.sort()
    return d_i_bar
```

```python
def laplacian_S(P,D):
    temp = np.identity(n+m)
    
    temp_D = np.zeros((n+m,n+m))
    for i in range(n+m):
        if D[i][i] != 0:
#             print(D[i][i])
            temp_D[i][i] = 1/math.sqrt(D[i][i])
        else:
            temp_D[i][i] = 0
    
    laplacian = temp - (temp_D @ P @ temp_D)
    return laplacian
```

```python
def update_A(S):
    num = 0
    denom = 0
    
    for j in range(m):
        num = 0
        denom = 0
        for i in range(n):
            num += S[i][j]*dataset[i]
            denom += S[i][j]
        print('num:' + str(num) + ' denom:' + str(denom))
        if denom != 0:
            A[j] = num/denom
    
    return A
            
```

```python
def outer_converge(S,A,d_x,count):
    return (count > K)
```

```python
def inner_converge(S,A,d_x,P,D,S_bar,F,counter):
    return (counter > K)
```

```python
count = 0
while (True):
    count += 1
    print('count:' + str(count))
    S = update_S(d_x)
    counter = 0
    while True:
        counter += 1
        print('count:' + str(count))
    #     print(S)
    #     print('rank:' + str(np.linalg.matrix_rank(S)))
        P = update_P(S)
        (D,D_U,D_V) = update_degrees(P)

        L = laplacian_S(P,D)
        print('rank:' + str(np.linalg.matrix_rank(L)))

        S_bar = update_S_bar(D_U,D_V)
        F = update_F(S_bar)
        d_bar = np.zeros((n,m))
        for i in range(n):
            d_i_bar = update_d_i_bar(F,D,i)
            d_bar[i] = d_i_bar
    #         print(d_i_bar)

        S = update_S(d_bar)
        print(S)
        print(S.shape)
        if inner_converge(S,A,d_x,P,D,S_bar,F,counter):
            break
    
    A = update_A(S)
    d_x = update_d_x(A)
    
    if (outer_converge(S,A,d_x,count)):
        break
```
